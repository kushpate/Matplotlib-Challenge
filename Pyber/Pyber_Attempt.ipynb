```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
ride_data = pd.merge(city_df,ride_df)
# Display the data table for preview
ride_data
```

```python
# Obtain the x and y coordinates for each of the three city types
urban_cities = ride_data[ride_data["type"] == 'Urban']
rural_cities = ride_data[ride_data["type"] == "Rural"]
suburban_cities = ride_data[ride_data["type"] == "Suburban"]

urban_ride_count = urban_cities.groupby(["city"])["ride_id"].count()
urban_avg_fare = urban_cities.groupby(["city"])["fare"].mean()
urban_driver_count = urban_cities.groupby(["city"])["driver_count"].count()

rural_ride_count= rural_cities.groupby(["city"])["ride_id"].count()
rural_avg_fare= rural_cities.groupby(["city"])["fare"].mean()
rural_driver_count= rural_cities.groupby(["city"])["driver_count"].count()

suburban_ride_count = suburban_cities.groupby(["city"])["ride_id"].count()
suburban_avg_fare = suburban_cities.groupby(["city"])["fare"].mean()
suburban_driver_count = suburban_cities.groupby(["city"])["driver_count"].count()
```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
plt.scatter(urban_ride_count, urban_avg_fare, s=20*urban_driver_count, c="coral", edgecolor="black", linewidths=1, marker="o", 
alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count, suburban_avg_fare, s=5*suburban_driver_count, c="skyblue", edgecolor="black", linewidths=1, 
marker="o", alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count, rural_avg_fare, s=20*rural_driver_count, c="gold", edgecolor="black", linewidths=1, marker="o", 
alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.xlim((0,40))

# Create a legend
lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="best", title="City Types", 
                  labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42, 35, "Note:\nCircle size correlates with driver count per city.")

# Show plot and save figure
plt.savefig('Pyber_Scatter.png')
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_percent = ride_data.groupby(['type']).sum()['fare'] / ride_data["fare"].sum()
# Build Pie Chart
plt.pie(type_percent, labels = ['Rural', 'Suburban', 'Urban'], colors = ['Yellow','SkyBlue','LightCoral'], startangle=140, 
        autopct='%1.1f%%')

plt.title('% of Total Fares by City Type')
# Save Figure
plt.savefig('Total_fares.png')
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_percents = 100 * (ride_data.groupby('type')['ride_id'].count()/ ride_data['ride_id'].count())
# Build Pie Chart
plt.pie(ride_percents, labels=['Rural', 'Suburban', 'Urban'], colors = ['Yellow','SkyBlue','LightCoral'],startangle=140, 
        autopct='%1.1f%%')
plt.title('Total Rides by City Type')
# Save Figure
plt.savefig('Total_rides.png')
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_percents =  100 *(ride_data.groupby('type')['driver_count'].count()/ride_data['driver_count'].count())
# Build Pie Charts
plt.pie(driver_percents, labels=['Rural', 'Suburban', 'Urban'], colors=['Yellow','SkyBlue','LightCoral'], startangle=140, 
       autopct='%1.1f%%')
plt.title('Total Drivers by City Type')
# Save Figure
plt.savefig('Total_Drivers_by_City_Type.png')
```
